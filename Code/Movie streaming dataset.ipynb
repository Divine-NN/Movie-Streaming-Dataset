### Removing duplicates 

```python
Moviestreaming.duplicated().sum()
```

```python
Moviestreaming.drop_duplicates(inplace = True)
Moviestreaming
```


### Cleaning of User name column

```python
Moviestreaming["User_Name"].value_counts()
```

```python
## solving the inconstitent text format issue
Moviestreaming["User_Name"] = Moviestreaming["User_Name"].str.strip().str.lower().str.title()
Moviestreaming["User_Name"]
```

```python
Moviestreaming["User_Name"].value_counts()
```

### Cleaning of Email_phone column

```python
Moviestreaming["Email_Phone"].value_counts()
```

```python
## splitting the email_phone column
Moviestreaming[["Email_Address", "Phone_Number"]] = Moviestreaming["Email_Phone"].str.split(",", expand= True)
Moviestreaming[["Email_Address", "Phone_Number"]]
```

### Cleaning country column 

```python
Moviestreaming["Country"].value_counts()
```

```python
Moviestreaming["Country"] = Moviestreaming["Country"].str.strip().str.lower().str.title()
Moviestreaming["Country"]
```

```python
Moviestreaming["Country"] = Moviestreaming["Country"].str.replace("Usa", "USA",case = True, regex= True)
Moviestreaming
```

### Cleaning Subscription_type column

```python
Moviestreaming["Subscription_Type"].value_counts()
```

```python
Moviestreaming["Subscription_Type"] = Moviestreaming["Subscription_Type"].str.strip().str.lower().str.title()
Moviestreaming["Subscription_Type"]
```

```python
Moviestreaming.info()
```

### Cleaning Genre_watched column

```python
Moviestreaming["Genre_Watched"] = Moviestreaming["Genre_Watched"].str.strip().str.lower().str.title()
Moviestreaming["Genre_Watched"]
```

### Dropping the transaction date column

```python
Moviestreaming = Moviestreaming.drop("Email_Phone", axis=1)
```

```python
Moviestreaming
```

```python
Moviestreaming.dropna(how = "all",inplace= True)
Moviestreaming
```

```python

Moviestreaming.isnull().sum()
```

### Converting null values to error

```python
Moviestreaming = Moviestreaming.fillna("UNKNOWN")
Moviestreaming
```

```python
Moviestreaming[["Watch_Time_Hours","Rating"]] = Moviestreaming[["Watch_Time_Hours","Rating"]].replace(["UNKNOWN", 0],pd.NA)
Moviestreaming
```

### Answering insight driven questions

```python
# What is the relationship between watch_time_hours and rating
Insight1 = Moviestreaming["Watch_Time_Hours"].corr(Moviestreaming["Rating"])
Insight1
```

```python
# what is average watch time per subscription_Type
Insight2 = Moviestreaming.groupby("Subscription_Type")["Watch_Time_Hours"].mean().reset_index(name = "Average_watchtimehour").sort_values(by="Average_watchtimehour",ascending=False)
Insight2
```

```python
# Which country has the highest users 
Insight3 = Moviestreaming.groupby("Country")["User_Name"].count().reset_index(name = "Number of users").sort_values(by="Number of users",ascending=False)
Insight3
```

```python
#What is the most watched genre
Insight4= Moviestreaming['Genre_Watched'].value_counts().to_frame(name='Count').reset_index()
Insight4
```

```python
# What is the most watched genre and average rating
Insight5 = Moviestreaming.groupby('Genre_Watched')['Rating'].mean().reset_index(name='Average rating').sort_values(by= 'Average rating', ascending=False)
Insight5
```

```python
# which subcription type has the  lowest rating
Insight6 = Moviestreaming.groupby("Subscription_Type")["Rating"].mean().reset_index(name = "Average_rating").sort_values(by = "Average_rating",ascending= False) 
Insight6
```

```python
# What is the number of users per subscription type
Insight7 = Moviestreaming.groupby("Subscription_Type")["User_Name"].count().reset_index(name = "Number of users").sort_values(by = "Number of users",ascending= False) 
Insight7
```
